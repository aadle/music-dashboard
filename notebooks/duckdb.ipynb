```python
import duckdb
from pathlib import Path
```

```python
data_dir = Path("../data/lastfm/listening")
db = duckdb.read_json(data_dir/"*.jsonl")
```

```python
db
```

```python
query = """
SELECT 
    COUNT(*) AS freq,
    track_name,
    album_name
FROM 
    db 
GROUP BY
    track_name,
    album_name
ORDER BY
   freq 
DESC
"""
duckdb.sql(query=query)
```

```python
query = """
SELECT 
    COUNT(*) AS freq,
    track_name,
    album_name
FROM 
    db 
WHERE 
    artist_name = 'Bladee'
GROUP BY
    track_name,
    album_name
ORDER BY
   freq 
DESC
"""
duckdb.sql(query=query)
```

```python
query = """
SELECT 
    COUNT(*) AS freq,
    album_name
FROM 
    db 
WHERE 
    artist_name = 'Bladee'
GROUP BY
    album_name
ORDER BY
   freq 
DESC
"""
duckdb.sql(query=query)
```

```python
query = """
SELECT 
    COUNT(*) AS freq,
    artist_name,
    to_timestamp(MIN(date_played_unix)) AT TIME ZONE 'UTC' as first_played,
    to_timestamp(MAX(date_played_unix)) AT TIME ZONE 'UTC' as last_played,
FROM 
    db 
GROUP BY
    artist_name
ORDER BY
   freq 
DESC
LIMIT 20
"""
duckdb.sql(query=query)
```
