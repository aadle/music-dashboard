# Listening habits over time

```python
import duckdb
import polars as pl
from pathlib import Path

data_dir = Path("../data/lastfm/listening")
db = duckdb.read_json(data_dir/"*.jsonl")
df = db.pl()

df = df.with_columns(
    pl.from_epoch(
        pl.col("date_played_unix"), time_unit="s"
    ).alias("track_played_utc")
)
```

```python
df.sort(pl.col("track_played_utc"), descending=True)
```

```python
import pendulum
last_week = pendulum.today().add(weeks=-1).naive()
jan_2026 = pendulum.datetime(2026, 1, 1).naive()
```

```python
# Filtering monthly
df.filter(
    pl.col("track_played_utc") >= jan_2026,
    pl.col("track_played_utc") < jan_2026.add(months=1),
)
```

```python
# Listening past 7 days
df.filter(
    pl.col("track_played_utc") >= pendulum.today().naive().add(weeks=-1),
    pl.col("track_played_utc") < pendulum.today().naive(),
)
```

```python
# Filtering daily
df.filter(
    pl.col("track_played_utc") >= last_week,
    pl.col("track_played_utc") < last_week.add(days=1),
)
```

## Listening streaks

```python
df_2025 = df.filter(
    pl.col("track_played_utc") >= pendulum.datetime(2025, 1, 1).naive(),
    pl.col("track_played_utc") < pendulum.datetime(2026, 1, 1).naive()
)
```

```python
dates_2025 = df_2025.select(
    pl.col("track_played_utc").dt.date().unique().alias("date")
)

listening_streak_2025 = dates_2025.with_columns(
    is_break = pl.col("date").diff() != pl.duration(days=1)
).with_columns(
    streak_id = pl.col("is_break").fill_null(True).cum_sum()
)

streak_counts_2025 = (
    listening_streak_2025
    .group_by(pl.col("streak_id"))
    .agg(
        start_date = pl.col("date").min(),
        end_date = pl.col("date").max(),
        streak_length = pl.len()
    )
)
streak_counts_2025
```

```python
def listening_streak(df:pl.DataFrame, year:int=2025):
    df = df.filter(
        pl.col("track_played_utc") >= pendulum.datetime(year, 1, 1).naive(),
        pl.col("track_played_utc") < pendulum.datetime(year+1, 1, 1).naive()
    )

    dates = df.select(
        pl.col("track_played_utc").dt.date().unique().alias("date")
    )

    listening_streak = dates.with_columns(
        is_break = pl.col("date").diff() != pl.duration(days=1) 
    ).with_columns( 
        streak_id = pl.col("is_break").fill_null(True).cum_sum()
    )

    streak_stats = (
        listening_streak
        .group_by(pl.col("streak_id"))
        .agg(
            start_date = pl.col("date").min(),
            end_date = pl.col("date").max(),
            streak_length = pl.len()
        )
    )
    streak_stats = streak_stats.sort(pl.col("streak_length"), descending=True)

    return streak_stats
```

```python
listening_streak(df=df, year=2021)
```

## Year-over-year listening trends

```python
tot_scrobbles_25 = df_2025.group_by(pl.col("track_played_utc").dt.date()).agg(total_scrobbles = pl.len())
tot_scrobbles_25 = tot_scrobbles_25.sort("track_played_utc", descending=True)
```

```python
import plotly.express as px
px.line(tot_scrobbles_25, "track_played_utc", "total_scrobbles")
```

```python
from scipy.signal import savgol_filter
total_scrobbles = tot_scrobbles_25.select(pl.col("total_scrobbles")).to_numpy().flatten()
tot_scrobbles_25 = tot_scrobbles_25.with_columns(filtered_scrobbles = savgol_filter(total_scrobbles, window_length=28, polyorder=3, axis=0))
tot_scrobbles_25
```

```python
import plotly.express as px
px.line(tot_scrobbles_25, "track_played_utc", [ "total_scrobbles", "filtered_scrobbles" ], title="2025")
```

## Most active listening hours of the day and days of the week

```python
df_2025 = df.filter(
    pl.col("track_played_utc") >= pendulum.datetime(2025, 1, 1).naive(),
    pl.col("track_played_utc") < pendulum.datetime(2026, 1, 1).naive()
)
```

```python
# peak hours (keep in mind hours in UTC, so +1h for actual timeslot)
(df_2025
    .group_by(pl.col("track_played_utc").dt.hour())
    .agg(scrobbles_by_hour = pl.len())
).sort("scrobbles_by_hour", descending=True)
```

```python
# peak days of the week
weekdays = {1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday",
            5: "Friday", 6: "Saturday", 7: "Sunday"}
(df_2025
    .group_by(pl.col("track_played_utc").dt.weekday().alias("weekday_idx"))
    .agg(scrobbles_by_weekday = pl.len())
    .sort("scrobbles_by_weekday", descending=True)
    .with_columns(weekday = pl.col("weekday_idx").replace(weekdays, default=None).alias("weekday"))
)
```
